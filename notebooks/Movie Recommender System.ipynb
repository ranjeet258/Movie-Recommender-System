```python
import ast  # here i am importing the ast module. This module will help me to convert the string representation of a list into an actual list. For example, if i have a string like this: "[1, 2, 3]", i can use ast.literal_eval() to convert it into a list like this: [1, 2, 3].
```

```python
def convert(text): # this function will take a string representation of a list and convert it into an actual list. For example, if i have a string like this: "[1, 2, 3]", this function will convert it into a list like this: [1, 2, 3].
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])   # here i am appending the name of the genre, keyword, cast member, or crew member to the list L. This is because i only want to consider the name of the genre, keyword, cast member, or crew member for my analysis. I don't care about the other information that is stored in the string (like the id or the popularity).
    return L
#string repre and actual list means that sometimes when we read data from a csv file, the list is stored as a string. For example, if we have a list like this: [1, 2, 3], it might be stored in the csv file as a string like this: "[1, 2, 3]". This function will take that string and convert it back into an actual list that we can use for our analysis.
```

```python
movies['genres']=movies['genres'].apply(convert) # here i am applying the convert function to the genres column. This will convert the string representation of the list into an actual list that we can use for our analysis.
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert) # here i am applying the convert function to the keywords column. This will convert the string representation of the list into an actual list that we can use for our analysis.
movies.head(1)
```

```python
def convert3(text):  # only give me the top 3 elements from the list. This is because some movies have a lot of cast members and crew members, and i only want to consider the top 3 for my analysis.
    L=[]
    count=0
    for i in ast.literal_eval(text):        
        if count < 3:
            L.append(i['name'])
            count+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)  # here i am applying the convert3 function to the cast column. This will convert the string representation of the list into an actual list that we can use for our analysis. It will also only give me the top 3 cast members for each movie.
```

```python
movies.head(1)
```

```python
def find_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(find_director)
movies.head(1)
```

```python
# let remove space from the names in the genres, keywords, cast, and crew columns. This is because when we will create a new column called 'tags' by combining all these columns, we want to make sure that there are no spaces in the names. This will help us to create a better tags column that we can use for our analysis. 
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
import nltk  # here i am importing the nltk library. This library will help me to do natural language processing tasks like stemming and lemmatization. Stemming is the process of reducing a word to its root form. For example, the word "running" will be reduced to "run". Lemmatization is the process of reducing a word to its base form. For example, the word "running" will be reduced to "run". I will use stemming in this project to reduce the words in the tags column to their root form. This will help me to create a better tags column that we can use for our analysis.
```

```python
from nltk.stem.porter import PorterStemmer  # here i am importing the PorterStemmer class from the nltk.stem.porter module. This class will help me to do stemming on the words in the tags column. Stemming is the process of reducing a word to its root form. For example, the word "running" will be reduced to "run". I will use stemming in this project to reduce the words in the tags column to their root form. This will help me to create a better tags column that we can use for our analysis.
ps = PorterStemmer()
```

```python
def stem(text): # i created this function to do stemming on the words in the tags column. This function will take a string as input, split it into words, stem each word using the PorterStemmer, and then join the stemmed words back into a single string. This will help me to create a better tags column that we can use for our analysis.
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)   # here i am applying the stem function to the tags column. This will stem each word in the tags column and then join the stemmed words back into a single string. This will help me to create a better tags column that we can use for our analysis.
```

```python
from sklearn.feature_extraction.text import CountVectorizer   #countvectorizer is a class in the sklearn library that will help me to convert the text data in the tags column into a matrix of token counts. This will allow me to use the tags for my analysis and build my movie recommender system.
cv = CountVectorizer(max_features = 5000, stop_words='english') # created a object cvd --- IGNORE ---
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()  #numpy array mei convert kiya #applying fit_transform() method to the tags column. This will convert the text data in the tags column into a matrix of token counts. The toarray() method will convert the sparse matrix into a numpy array that we can use for our analysis.
```

now we have a matrix of token counts for each movie in the tags column. Each row in the matrix represents a movie, and each column represents a word in the tags. The value in each cell represents the number of times that word appears in the tags for that movie. For example, if the value in the cell is 3, it means that the word appears 3 times in the tags for that movie. If the value is 0, it means that the word does not appear in the tags for that movie.


```python
vector[0] # there will many zeros(Sparsity )as only few words available for one movies # in short this is the vector representation of the tags column for the first movie in the dataframe. Each element in the vector represents the count of a particular word in the tags column for that movie. Since there are many words in the tags column, there will be many zeros in the vector because most movies will not have all the words in their tags. This is known as sparsity, and it is a common issue when working with text data.
```

```python
from sklearn.metrics.pairwise import cosine_similarity  # here i am importing the cosine_similarity function from the sklearn.metrics.pairwise module. This function will help me to calculate the cosine similarity between the vector representations of the tags column for each movie. The cosine similarity is a measure of similarity between two vectors. It ranges from -1 to 1, where 1 means that the vectors are identical, 0 means that the vectors are orthogonal (i.e., they have no similarity), and -1 means that the vectors are opposite (i.e., they have maximum dissimilarity). I will use the cosine similarity to find similar movies based on their tags.
```

```python
similarity = cosine_similarity(vector)  #calcu similarity between the vector representations of the tags column for each movie. This will give me a matrix of cosine similarity values, where each element (i, j) in the matrix represents the cosine similarity between the tags of movie i and movie j. I can then use this matrix to find similar movies based on their tags. For example, if I want to find movies that are similar to a particular movie, I can look at the corresponding row in the similarity matrix and find the movies that have high cosine similarity values with that movie.
```

```python
similarity[0]   # distance b/w first movie with each movie, diagonal will always be one , similarity[0][0] # distance b/w first movie with itself, it will always be one because the cosine similarity of a vector with itself is always 1. This is because the angle between the vector and itself is 0 degrees, and the cosine of 0 degrees is 1. Therefore, the diagonal elements of the similarity matrix will always be 1, indicating that each movie is perfectly similar to itself.
```

```python
similarity # this will give me the entire similarity matrix. Each element (i, j) in the matrix represents the cosine similarity between the tags of movie i and movie j. I can use this matrix to find similar movies based on their tags. For example, if I want to find movies that are similar to a particular movie, I can look at the corresponding row in the similarity matrix and find the movies that have high cosine similarity values with that movie.
```

```python
# list(enumerate(similarity[6])) makes sure indexing
#key=lambda x:x[1] means that we want to sort the list based on the second element of the tuple, which is the cosine similarity value. This will help us to sort the movies based on their similarity to the 7th movie.
# reverse = True means that we want to sort the list in descending order based on the cosine similarity values. This will help us to get the most similar movies at the top of the list.
sorted(list(enumerate(similarity[6 ])), reverse = True,key=lambda x:x[1])[1:6] #  sorting the list of cosine similarity values for the 7th movie in descending order. The enumerate() function is used to get the index of each movie along with its cosine similarity value. The sorted() function is used to sort the list based on the cosine similarity values (which is the second element of the tuple). The reverse=True argument is used to sort in descending order. Finally, [1:6] is used to get the top 5 similar movies (excluding the first one which is the movie itself). This will give me a list of tuples, where each tuple contains the index of a similar movie and its cosine similarity value with the 7th movie.
```

```python
# let make funtion for recommendation so that we can easily get the similar movies by just passing the movie name as an argument to the function. This will make it easier for us to use the movie recommender system and get similar movies based on the tags.
def recommend(movie):  # this function will take a movie title as input and recommend 5 similar movies based on the cosine similarity of the tags. The function will first find the index of the movie in the new_df dataframe, then it will get the corresponding row from the similarity matrix, sort the values in descending order, and finally return the titles of the top 5 similar movies (excluding the first one which is the movie itself).
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```
